1.3 Напишите функцию, которая по `username` ищет человека и выводит информацию о нем. Если пользователь с заданным `username` отсутствует, возбудите исключение `ValueError`

````python
def find_user_info(name, filename):
    with open(filename) as file:
        users = json.load(file)

    for user in users:
        if user.get('name') == name:
            return user
    return None

filename = 'contributors_sample.json'

name = input("введите имя: ")
user_info = find_user_info(name, filename)

if user_info is not None:
    print(user_info)
else:
    print(f"{name}: нет такого")
````

### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
# Загрузка данных из JSON файла
with open('contributors_sample.json') as f:
    contributors_data = json.load(f)

# Инициализация словаря для хранения информации о должностях
positions_dict = {}

# Обработка каждого человека в данных
for person in contributors_data:
    # Перебираем каждую должность для текущего человека
    for job in person['jobs']:
        # Если должность уже есть в словаре, добавляем имя пользователя в список
        if job in positions_dict:
            positions_dict[job].append(person['username'])
        # Если должности нет в словаре, создаем новую запись
        else:
            positions_dict[job] = [person['username']]

# Форматирование словаря для вывода
formatted_output = ""
for job, usernames in positions_dict.items():
    # Добавляем заголовок должности
    formatted_output += f"\n{job}:\n"
    # Перебираем имена пользователей для текущей должности
    for username in usernames:
        # Добавляем каждое имя пользователя с префиксом "-"
        formatted_output += f"- {username}\n"

# Вывод отформатированного текста
print(formatted_output)
````

2.2 Сохраните результаты в файл `job_people.pickle` и в файл `job_people.json` с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент `indent`.

````python
# Загрузка данных из JSON файла
with open('contributors_sample.json', 'r') as f:
    contributors_data = json.load(f)

# Инициализация словаря для хранения информации о должностях
positions_dict = {}

# Обработка каждого человека в данных
for person in contributors_data:
    for job in person['jobs']:
        # Если должность уже есть в словаре, добавляем имя пользователя в список
        if job in positions_dict:
            positions_dict[job].append(person['username'])
        # Если должности нет в словаре, создаем новую запись
        else:
            positions_dict[job] = [person['username']]

# Сохранение результатов в файл job_people.pickle
with open('job_people.pickle', 'wb') as f:
    pickle.dump(positions_dict, f)

# Сохранение результатов в файл job_people.json с аргументом indent
with open('job_people.json', 'w') as f:
    json.dump(positions_dict, f, indent=4)

# Сообщение о завершении и просьба сравнить размеры файлов
print("Результаты сохранены в job_people.pickle и job_people.json.")
print("Пожалуйста, сравните их объемы, чтобы увидеть разницу в размерах.")
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
# Загрузка данных из XML файла
tree = ET.parse('steps_sample.xml')
root = tree.getroot()

# Инициализация словаря для хранения шагов по каждому рецепту
steps_dict = {}

# Обработка каждого рецепта в данных
for i, recipe in enumerate(root.findall('recipe'), start=1):
    # Генерируем уникальный идентификатор для рецепта
    recipe_id = f"recipe_{i}"
    # Получаем список шагов для текущего рецепта
    steps = [step.text for step in recipe.findall('step')]
    # Добавляем список шагов в словарь по соответствующему идентификатору рецепта
    steps_dict[recipe_id] = steps

# Вывод словаря для проверки
for recipe_id, steps in steps_dict.items():
    print(f"Рецепт {recipe_id}: {steps}")

# Сохранение словаря в файл steps_sample.json
with open('steps_sample.json', 'w') as f:
    # Записываем словарь в формате JSON с отступами для читаемости
    json.dump(steps_dict, f, indent=4)
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
# Загрузка данных из XML файла
tree = ET.parse('steps_sample.xml')
root = tree.getroot()

# Инициализация словаря для хранения количества шагов в рецептах
steps_count_dict = {}

# Обработка каждого рецепта в данных
for i, recipe in enumerate(root.findall('recipe'), start=1):
    recipe_id = f"recipe_{i}"
    steps_count = len(recipe.findall('step'))
    
    # Если количество шагов уже есть в словаре, добавляем ID рецепта в соответствующий список
    if steps_count in steps_count_dict:
        steps_count_dict[steps_count].append(recipe_id)
    # Если количество шагов отсутствует в словаре, создаем новую запись
    else:
        steps_count_dict[steps_count] = [recipe_id]

# Вывод словаря для проверки
for steps, recipes in steps_count_dict.items():
    print(f"Количество шагов: {steps}, Рецепты: {recipes}")

# Сохранение словаря в файл steps_counts_sample.json
with open('steps_counts_sample.json', 'w') as f:
    json.dump(steps_count_dict, f, indent=4)
````

3.4 Загрузите данные из файла `recipes_sample.csv` (__ЛР2__) в таблицу `recipes`. Для строк, которые содержат пропуски в столбце `n_steps`, заполните этот столбец на основе файла  `steps_sample.xml`. Строки, в которых столбец `n_steps` заполнен, оставьте без изменений.

````python
# Загрузка данных из XML файла
tree = ET.parse('steps_sample.xml')  # Парсинг XML файла и создание объекта дерева
root = tree.getroot()  # Получение корневого элемента дерева

# Инициализация словаря для хранения шагов по каждому рецепту
steps_dict = {}

# Обработка каждого рецепта в данных
for i, recipe in enumerate(root.findall('recipe'), start=1):
    # Генерация уникального идентификатора для рецепта
    recipe_id = f"recipe_{i}"
    # Получение списка шагов для текущего рецепта
    steps = [step.text for step in recipe.findall('step')]
    # Добавление списка шагов в словарь по соответствующему идентификатору рецепта
    steps_dict[recipe_id] = steps

# Вывод словаря для проверки
for recipe_id, steps in steps_dict.items():
    print(f"Рецепт {recipe_id}: {steps}")

# Сохранение словаря в файл steps_sample.json
with open('steps_sample.json', 'w') as f:
    # Запись словаря в формате JSON с отступами для читаемости
    json.dump(steps_dict, f, indent=4)
````

3.5 Проверьте, содержит ли столбец `n_steps` пропуски. Если нет, то преобразуйте его к целочисленному типу и сохраните результаты в файл `recipes_sample_with_filled_nsteps.csv`

````python
import pandas as pd

# Загрузка данных из файла recipes_sample.csv в таблицу recipes
recipes = pd.read_csv('recipes_sample.csv')

# Проверка наличия пропусков в столбце n_steps
if recipes['n_steps'].isnull().any():
    print("Столбец n_steps содержит пропуски.")
else:
    print("Столбец n_steps не содержит пропусков.")
    
    # Преобразование столбца n_steps к целочисленному типу
    recipes['n_steps'] = recipes['n_steps'].astype(int)
    
    # Сохранение результатов в файл recipes_sample_with_filled_nsteps.csv
    recipes.to_csv('recipes_sample_with_filled_nsteps.csv', index=False)
    print("Результаты сохранены в файл recipes_sample_with_filled_nsteps.csv")
````
