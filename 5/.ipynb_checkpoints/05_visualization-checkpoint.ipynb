1\. В файле `average_ratings.npy` содержится информация о среднем рейтинге 3 рецептов за период с 01.01.2019 по 30.12.2021. При помощи пакета `matplotlib` в _одной системе координат_ (на одной картинке) изобразите три временных ряда, соответствующих средним рейтингам этих рецептов. 

По горизонтальной оси располагается номер дня (0, 1, 2, ...), по вертикальной - средний рейтинг рецептов в этот день. 

Названия рецептов и их индексы в файле `average_ratings.npy`:
```
0: waffle iron french toast
1: zwetschgenkuchen bavarian plum cake
2: lime tea
```

Результатом работы является визуализация, на которой:
* добавлена подпись горизонтальной оси с текстом "Номер дня"
* добавлена подпись вертикальной оси с текстом "Средний рейтинг"

````python
# Загружаем данные из файла в формате numpy
ratings = np.load('data/average_ratings.npy')  # Загружаем массив данных о средних рейтингах из файла 'average_ratings.npy'

# Отображаем загруженные данные
ratings
````

````python

# Создаем массив значений x, представляющих дни, от 0 до длины первого набора данных в ratings
x = np.arange(0, len(ratings[0]))

# Извлекаем данные для первого рецепта
y_1 = ratings[0]
# Извлекаем данные для второго рецепта
y_2 = ratings[1]
# Извлекаем данные для третьего рецепта
y_3 = ratings[2]

# Строим график для первого рецепта и добавляем метку
plt.plot(x, y_1, label="waffle iron french toast")
# Строим график для второго рецепта и добавляем метку
plt.plot(x, y_2, label="zwetschgenkuchen bavarian plum cake")
# Строим график для третьего рецепта и добавляем метку
plt.plot(x, y_3, label="lime tea")

# Добавляем подпись для оси y
plt.ylabel('Средний рейтинг')
# Добавляем подпись для оси x
plt.xlabel('Номер дня')
# Добавляем заголовок графика
plt.title('Изменение среднего рейтинга трех рецептов')
# Добавляем легенду в верхний правый угол с выносом
plt.legend(loc="upper right", bbox_to_anchor=(1.7, 1))

# Отображаем график
plt.show()
````

5\. Назовем рецепты короткими, если они занимают строго меньше 5 минут; средними, если они занимают от 5 до 50 минут (не включая 50), и длинными, если они занимают от 50 минут и больше. Сгруппируйте все рецепты по данному признаку и для каждой группы посчитайте 2 величины: среднее количество шагов рецептов в группе и размер группы. При помощи методов структур `pandas` постройте столбчатую диаграмму, где каждый столбец означает группу (короткие, средние или длинные рецепты), а высота столбца обозначает среднее количество шагов рецептов в группе. Рядом по горизонтали разместите круговую диаграмму, на которой отображены размеры каждой из групп.

Добавьте следующие подписи:
* по горизонтальной оси под столбчатой диаграммой напишите "Группа рецептов"
* по вертикальной оси слева от столбчатой диаграммы напишите "Средняя длительность"
* над круговой диаграммой напишите "Размеры групп рецептов"

````python

# Фильтруем рецепты по длительности
recipes_small = recipes[recipes['minutes'] < 5]
recipes_average = recipes[(recipes['minutes'] >= 5) & (recipes['minutes'] < 50)]
recipes_long = recipes[recipes['minutes'] >= 50]

# Объединяем отфильтрованные данные в один DataFrame
df = pd.concat([recipes_small, recipes_average, recipes_long])

# Создаем фигуру и оси для двух графиков, расположенных в одной строке
fig, axs = plt.subplots(1, 2, figsize=(15, 8), constrained_layout=True)

# Первый график - столбчатая диаграмма средней длительности шагов по группам рецептов
ax = axs[0]
# Вычисляем среднюю длительность шагов для каждой группы рецептов
mean_steps = [
    recipes_small['n_steps'].mean(),
    recipes_average['n_steps'].mean(),
    recipes_long['n_steps'].mean()
]
# Строим столбчатую диаграмму
ax.bar(["Короткие", "Средние", "Длинные"], mean_steps, color=['blue', 'yellow', 'green'])
# Устанавливаем подписи для осей
ax.set_xlabel("Группа рецептов", fontsize=18)
ax.set_ylabel("Средняя длительность", fontsize=18)

# Второй график - круговая диаграмма размеров групп рецептов
ax = axs[1]
# Вычисляем количество рецептов в каждой группе
y = [len(recipes_small), len(recipes_average), len(recipes_long)]
# Строим круговую диаграмму
ax.pie(y, labels=["Короткие", "Средние", "Длинные"], autopct='%1.1f%%', colors=['blue', 'green', 'yellow'])
# Устанавливаем подпись для оси x
ax.set_xlabel('Размеры групп рецептов', fontsize=18)

# Устанавливаем общий заголовок для фигуры
plt.suptitle("Размеры групп рецептов", fontsize=18)

# Отображаем графики
plt.show()
````

Добавьте в ячейки тепловой карты подписи (значения к-та корреляции). Измените цветовую палитру на `YlOrRd`. 

Добавьте заголовок рисунка "Корреляционная матрица числовых столбцов таблиц recipes и reviews"

````python
all_join = recipes.join(reviews)
correlation_matrix = all_join[['minutes', 'n_steps', 'n_ingredients', 'rating']].corr()

plt.figure(figsize=(15, 9))
plt.suptitle("Корреляционная матрица числовых столбцов таблиц recipes и reviews", fontsize=18)

ax0 = plt.axes()
sns.heatmap(correlation_matrix, ax=ax0, cmap=sns.color_palette('YlOrRd'), annot=True)
````
