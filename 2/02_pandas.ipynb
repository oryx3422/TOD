```python
# Преобразование названий столбцов DataFrame df_recipes в заголовочный формат (каждое слово с заглавной буквы)
# Для этого используется генератор списка (list comprehension), который проходит по каждому элементу списка, полученного из названий столбцов, и применяет метод title() для каждого элемента
# Результат сохраняется в новом списке new_names
new_names = [i.title() for i in df_recipes.columns.values.tolist()]

# Создание копии DataFrame df_recipes с сохранением глубокого копирования
# Для этого используется метод copy(), который создает новый объект DataFrame с теми же данными, что и исходный DataFrame, но с независимым набором данных
# Аргумент deep=True указывает на глубокое копирование, чтобы все вложенные объекты также были скопированы
df_recipes1 = df_recipes.copy(deep=True)

# Присвоение новых названий столбцов DataFrame df_recipes1
# Для этого используется атрибут columns, который содержит названия столбцов DataFrame, и присваивание новых названий из списка new_names
df_recipes1.columns = new_names

# Вывод первых нескольких строк DataFrame df_recipes1 с обновленными названиями столбцов
# Это делается с помощью метода head(), который выводит первые несколько строк DataFrame
df_recipes1.head()
```

3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
# Определение функции changes, которая принимает строку x в качестве аргумента и возвращает количество слов в этой строке
# Для этого используется метод split(), который разбивает строку на слова по пробелам, и функция len(), которая возвращает количество элементов в полученном списке слов
def changes(x):
    return len(x.split())

# Применение функции changes к столбцу 'name' DataFrame df_recipes
# Для этого используется метод apply(), который применяет функцию к каждому элементу столбца 'name'
# Результат сохраняется в новом столбце 'name_word_count'
df_recipes['name_word_count'] = df_recipes.name.apply(changes)

# Вывод первых нескольких строк DataFrame df_recipes с новым столбцом 'name_word_count'
# Это делается с помощью метода head(), который выводит первые несколько строк DataFrame
df_recipes.head()
```

### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
# Группировка DataFrame df_recipes по столбцу 'contributor_id' с помощью метода groupby()
# Для этого используется метод groupby(), который группирует строки по уникальным значениям в столбце 'contributor_id'
# Далее вызывается метод size(), который подсчитывает количество строк в каждой группе
# Результат группировки сохраняется в DataFrame recipes1
recipes1 = df_recipes.groupby('contributor_id').size().reset_index(name='recipe_count')

# Выбор строки (или строк), где количество рецептов (столбец 'recipe_count') равно максимальному значению в этом столбце
# Для этого сначала происходит сравнение значений в столбце 'recipe_count' с максимальным значением с помощью условного выражения recipes1['recipe_count'] == recipes1['recipe_count'].max()
# Результат сравнения - булева серия - используется для индексации DataFrame, чтобы выбрать строки с максимальным значением
max_recipe_contributors = recipes1[recipes1['recipe_count'] == recipes1['recipe_count'].max()]

# Вывод строк DataFrame max_recipe_contributors, которые содержат информацию о самых активных авторах рецептов
max_recipe_contributors
```

### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
# 1. Объединяем таблицы по столбцам 'id' и 'recipe_id'
merged1 = pd.merge(df_recipes, df_reviews, left_on='id', right_on='recipe_id')

# 2. Отбираем только те строки, где есть отзывы
merged1 = merged1.dropna(subset=['rating'])

# 3. Выбираем нужные столбцы
result_df = merged1[['id', 'name', 'user_id', 'rating']]
result_df
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
# Создание глубокой копии DataFrame merged1 с сохранением глубокого копирования
# Для этого используется метод copy(), который создает новый объект DataFrame с теми же данными, что и исходный DataFrame, но с независимым набором данных
# Аргумент deep=True указывает на глубокое копирование, чтобы все вложенные объекты также были скопированы
merged2 = merged1.copy(deep=True)

# Группировка DataFrame merged2 по столбцу 'id' с помощью метода groupby()
# Для этого используется метод groupby(), который группирует строки по уникальным значениям в столбце 'id'
# Далее вызывается метод count(), который подсчитывает количество строк (оценок) в каждой группе
# Результат подсчета сохраняется в новом столбце 'review_count'
merged2['review_count'] = merged2.groupby('id').count()['rating']

# Удаление строк с пропущенными значениями в столбце 'review_count' с помощью метода dropna()
# Для этого используется метод dropna(), который удаляет строки с пропущенными значениями в указанных столбцах (в данном случае - 'review_count')
# Затем выбираются только нужные столбцы 'id', 'name' и 'review_count' с помощью метода loc[]
# Новый DataFrame содержит информацию о количестве оценок для каждого рецепта и его названии
# Наконец, столбец 'id' переименовывается в 'recipes_id' с помощью метода rename()
merged2 = merged2.dropna(subset=['review_count']).loc[:, ['id', 'name', 'review_count']].rename(columns={'id': 'recipes_id'})

# Вывод первых нескольких строк нового DataFrame merged2
# Это делается с помощью метода head(), который выводит первые несколько строк DataFrame
merged2.head()
```
